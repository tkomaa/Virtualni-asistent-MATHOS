```python
import pandas as pd

df = pd.read_csv("baza_faq.csv", encoding="utf-8")
print(df.shape)
df.head()
```

```python
required = {"id","question","intent","answer_id","answer","source","split"}
print("Missing:", required - set(df.columns))
print(df["split"].value_counts())
print("\nIntents:", df["intent"].nunique())
```

```python
train_df = df[df["split"]=="train"].reset_index(drop=True)
dev_df   = df[df["split"]=="dev"].reset_index(drop=True)
test_df  = df[df["split"]=="test"].reset_index(drop=True)

print(len(train_df), len(dev_df), len(test_df))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report

vectorizer = TfidfVectorizer(ngram_range=(1,2), min_df=1)
X_train = vectorizer.fit_transform(train_df["question"])
X_dev   = vectorizer.transform(dev_df["question"])
X_test  = vectorizer.transform(test_df["question"])

clf = LogisticRegression(max_iter=2000)
clf.fit(X_train, train_df["intent"])

dev_pred = clf.predict(X_dev)
test_pred = clf.predict(X_test)

print("DEV accuracy:", accuracy_score(dev_df["intent"], dev_pred))
print("TEST accuracy:", accuracy_score(test_df["intent"], test_pred))
print("\nTEST report:\n", classification_report(test_df["intent"], test_pred))
```
